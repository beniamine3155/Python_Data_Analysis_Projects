## Aggregating the rows by the country

```python
aggregating = data1.groupby('Country/Region').sum()
```

```python
aggregating.sample(5)
```

```python
aggregating.shape
```

## Visualizing data related to a country for example Pakistan

```python
aggregating.loc['Iran'].plot()
aggregating.loc['Bhutan'].plot()
aggregating.loc['Pakistan'].plot()
plt.legend()
```

```python
aggregating.loc['Portugal'].plot()
```

```python
aggregating.loc['Portugal'].diff().plot()
```
